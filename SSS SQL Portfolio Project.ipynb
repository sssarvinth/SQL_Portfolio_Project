```python
## Scenario 33:- Create a case statement with a set containing the employee number, first name, and last name of all employees with a number higher than 100005. Create a fourth column in the query, indicating whether this employee is also a manager, according to the data provided in the dept_manager table, or a regular employee.
sql='''
SELECT

    e.emp_no,

    e.first_name,

    e.last_name,

    CASE

        WHEN dm.emp_no IS NOT NULL THEN 'Manager'

        ELSE 'Employee'

    END AS is_manager

FROM

    employees e

        LEFT JOIN

    dept_manager dm ON dm.emp_no = e.emp_no

WHERE

    e.emp_no > '10005'; 
'''
df = pd.read_sql_query(sql,conn)
df.head(100)
```
